#### **Mortgage 1:**
* Time: 30 years
* Rate: 4%
* Principal Amount: $1,000,000

```python
principal = 1000000
rate = 0.04
period = 12*30 # 30 years
remainingprincipal = 0
frame = []

for p in range(period+1):
    openningprincipal = remainingprincipal

    if p == 0:
        cashflow = 0
        interestpaid = 0
        principalpaid = cashflow-interestpaid
        remainingprincipal = principal
        frame.append([p, principal, rate, cashflow,
                        interestpaid, principalpaid, remainingprincipal])
    else:
        if period == 0:
            cashflow = openningprincipal
        else:
            cashflow = (principal * rate/12) / \
                (1 - 1/((1+rate/12)**(period)))
        interestpaid = openningprincipal*rate/12
        principalpaid = cashflow-interestpaid
        remainingprincipal = openningprincipal-principalpaid
        frame.append([p, openningprincipal, rate, cashflow,
                        interestpaid, principalpaid, remainingprincipal])

    mortgage1 = pd.DataFrame(frame, columns=["Month", "Opening Principal Balance", "Rates", "Cashflow",
                                    "Interest Paid", "Principal Paid", "Closing Princial Balance"])
    mortgage1.set_index("Month", inplace=True)

    mortgage1.Rates = (mortgage1.Rates*100).map('{:,.2f}%'.format)

print("Total Interest Paid: ${:#,.2f}".format(mortgage1['Interest Paid'].sum()))
mortgage1
```

#### **Mortgage 3:**
* Time: 30 years
* Rate: 7-1 Adjustable
* Principal Amount: $1,000,000

```python
file = pd.read_csv("https://docs.google.com/spreadsheets/d/e/"
                   "2PACX-1vTa_hFe6PiKrpkNiEUswWvkrUxXNP8OocOD7"
                   "qyviwIBXiC-oieH9nszD1j_wQHVpvoeTou4VnK5rHgI/"
                   "pub?gid=387420101&single=true&output=csv", skiprows=6)
rates = file['Rates'].dropna().str.replace('%','').unique().astype('float')/100
rates
```

```python
principal = 1000000
rate = rates[0]
interest = principal * rate/12
principals=[]
close_principals = []
interests = []
principal_paid = []
rates_list = []
cashflows = []
mortgage3 = pd.DataFrame()
```

```python
payment=npf.pmt(rates[0]/12, 30*12, -principal)
```

```python
for i in range(0, 7*12):
    principals.append(principal)
    interest = principal * rate/12
    interests.append(interest)
    paid = payment-interest
    principal_paid.append(paid)
    principal = principal - (paid)
    close_principals.append(principal)
    rates_list.append(rate)
    cashflows.append(payment)
```

```python
t = 0
for r in rates[1:]:
    payment=npf.pmt(r/12, 23*12-t*12, -principal)
    for i in range(0, 12):
        principals.append(principal)
        interest = principal * r/12
        interests.append(interest)
        paid = payment-interest
        principal_paid.append(paid)
        principal = principal - (paid)
        close_principals.append(principal)
        rates_list.append(r)
        cashflows.append(payment)
    t = t+1
```

```python
mortgage3['Opening Principal Balance'] = principals
mortgage3['Rates'] = rates_list
mortgage3['Cashflow'] = cashflows
mortgage3['Interest Paid'] = interests
mortgage3['Principal Paid'] = principal_paid
mortgage3['Closing Principal Balance'] = close_principals
mortgage3['Month'] = mortgage3.index+1
mortgage3.set_index('Month', inplace=True)
mortgage3['Rates'] = (mortgage3['Rates']*100).map('{:,.2f}%'.format)
```
